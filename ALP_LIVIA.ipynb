##Operadores

divisão /

multiplicação *

resto da divisão %

```python
numero1 = 10
numero2 = 5

print("A soma é igual a: ", numero1 + numero2)
print("A subtração é igual a: ", numero1 - numero2)
print("A divisão é igual a: ", numero1 / numero2)
print("A multiplicação é igual a: ", numero1 * numero2)
print("A resto da divisão é igual a: ", numero1 % numero2)
print("A divisão inteira é igual a: ", numero1 // numero2)
print("A exponenciação é igual a: ", numero1 ** numero2)
```

##Tipos de dados

##Estrutura Condicional

if - se

elif - senão se

else - senão

```python
nota = float(input("Digite a sua nota: "))

if(nota >= 7):
  print("Aprovado")
elif(nota >= 5):
  print("Recuperação")
else:
  print("Reprovado")
```

4 - Crie um algorítmo que leia a sua idade e dependendo da idade mostre as frases na tela:

até 12 anos - Você é uma Criança!

13 - 18 - Você é Adolescente!

19 - 60 - Você é Adulto!

acima de 60 - Você é Idoso!



```python
idade = int(input("Digite sua idade:"))
if(idade <= 12):
  print("Você é uma criança")
elif(idade > 12 and idade <= 18):
  print("Você é um adolescente")
elif(idade > 18 and idade <= 60):
  print("Você é um adulto")
else:
  print("Você é um idoso")
```
